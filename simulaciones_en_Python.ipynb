```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

<center>
    <h1>Crecimiento poblacional</h1>
</center>

Simulemos la ecuación más sencilla de crecimiento de una población:

$$ \frac{dx}{dt} = ax $$

```python
def sistema(x, t, a):
    dx_dt = a*x # Definimos la ecuación diferencial
    return dx_dt # La función debe dar como valor de retorno las ecuaciones diferenciales

x0 = 10 # Definimos un valor inicial para la población
t = np.linspace(0, 5, 1000) # Creamos el vector tiempo
a = np.log(2) # Definimos el parámetro a
sol = odeint(sistema, x0, t, args=(a,)) # Resolvemos el sistema por integración numérica

# Grafiquemos la solución
plt.plot(t, sol)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```

<center>
    <a id='ryj'></a>
    <h1>Romeo y Julieta</h1>
</center>
Simulemos el sistema de Romeo y Julieta:

$$ \frac{dR}{dt} = J $$

$$ \frac{dJ}{dt} = -R $$

```python
def ryj(y, t):
    r, j = y # y es el vector de variables de estado, debemos "desempaquetarlo"
    dr_dt = j
    dj_dt = -r
    return dr_dt, dj_dt
```

```python
t = np.linspace(0,15,1000)
y0 = [1 , 0] # Definimos las condiciones iniciales para r y j

sol = odeint(ryj, y0, t)
```

Grafiquemos los resultados:

```python
fig, ax = plt.subplots(1, 2, figsize=(12,6))
ax[0].plot(t, sol[:,0], label='Romeo')
ax[0].plot(t, sol[:,1], label='Julieta')
ax[0].legend(loc='best')
ax[0].set_title('Evolución del sistema')
ax[0].set(xlabel='Tiempo')

ax[1].plot(sol[:,0], sol[:,1])
ax[1].set(xlabel='Romeo', ylabel='Julieta')
ax[1].set_title('Romeo vs. Julieta')
ax[1].grid()
```
