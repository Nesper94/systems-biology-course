````python
import igraph as ig
````

````python
g = ig.Graph()
````

````python
print(g)
````

````python
g.add_vertices(3)
````

````python
print(g)
````

````python
g.add_edges([(0,1), (1,2)])
````

````python
print(g)
````

````python
g.add_edges([(2, 0)])

g.add_vertices(3)

g.add_edges([(2, 3), (3, 4), (4, 5), (5, 3)])
````

````python
g.get_edgelist()
````

````python
g.degree()
````

````python
g.degree(2)
````

## Medidas de centralidad
